```python
import pandas as pd
import numpy as np

# Create DataFrame
df_sales = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', end='2023-04-09'),
    'Region': np.random.choice(['East', 'West', 'North', 'South'], 99),
    'Product': np.random.choice(['Laptop', 'Mouse', 'Keyboard', 'Monitor'], 99),
    'Sales': np.random.randint(100, 1001, 99),
    'Quantity': np.random.randint(1, 11, 99)
})

# Introduce 5 NaN values in Sales
sales_nan_index = np.random.choice(df_sales.index, 5, replace=False)
df_sales.loc[sales_nan_index, 'Sales'] = np.nan

# Introduce 3 NaN values in Quantity
quantity_nan_index = np.random.choice(df_sales.index, 3, replace=False)
df_sales.loc[quantity_nan_index, 'Quantity'] = np.nan

# Display first 10 rows
print(df_sales.head(10))
```

```python
# --------------------------------------------------
# Task 3: Indexing and Slicing in DataFrames
# --------------------------------------------------

print("\n--- Task 3: Indexing and Slicing in DataFrames ---")

# Use 'df_sales' from Task 2.


# --------------------------------------------------
# 3.1 Select only the 'Region' and 'Sales' columns
# --------------------------------------------------

# Double square brackets [[]] are used to select multiple columns
df3_1 = df_sales[['Region', 'Sales']]

print("\n3.1 Region and Sales columns:")
print(df3_1.head())


# --------------------------------------------------
# 3.2 Select rows where Product is 'Laptop'
# --------------------------------------------------

# Select rows where the Product column value is 'Laptop'
df3_2 = df_sales[df_sales['Product'] == 'Laptop']

print("\n3.2 Rows where Product is Laptop:")
print(df3_2.head())


# --------------------------------------------------
# 3.3 Select rows where Region is 'East'
# AND Sales are greater than 500
# --------------------------------------------------

# '&' represents AND operation
# Each condition must be enclosed in parentheses

df3_3 = df_sales[
    (df_sales['Region'] == 'East') &
    (df_sales['Sales'] > 500)
]

print("\n3.3 Region = East and Sales > 500:")
print(df3_3)


# --------------------------------------------------
# 3.4 Select the Sales value for the first Monitor
# using .loc and .iloc
# --------------------------------------------------

# .loc selects rows and columns by label
# .iloc[0] selects the first matching row

val3_4 = df_sales.loc[
    df_sales['Product'] == 'Monitor',
    'Sales'
].iloc[0]

print("\n3.4 First Monitor Sales value:")
print(val3_4)
```

```python
# Task 4: Data Cleaning (Handling Missing Values)

print("\n--- Task 4: Data Cleaning (Handling Missing Values) ---")

# 4.1 Count missing values in each column
missing_counts4_1 = df_sales.isnull().sum()

print("\nMissing values in each column:")
print(missing_counts4_1)


# 4.2 Fill missing Sales values with the mean
df4_2_filled = df_sales.copy()

sales_mean = df4_2_filled['Sales'].mean()

df4_2_filled['Sales'] = df4_2_filled['Sales'].fillna(sales_mean)

print("\nDataFrame after filling missing Sales values:")
print(df4_2_filled.head())


# 4.3 Drop rows with any missing values
df4_3_dropped = df_sales.dropna()

print("\nDataFrame after dropping rows with missing values:")
print(df4_3_dropped.head())
```

```python
# --------------------------------------------------
# Task 5: Data Manipulation (Columns, Sorting, Grouping)
# --------------------------------------------------

print("\n--- Task 5: Data Manipulation (Columns, Sorting, Grouping) ---")

# Use the original df_sales DataFrame

# --------------------------------------------------
# 5.1 Create a new column 'Total_Revenue'
# Total_Revenue = Sales * Quantity
# --------------------------------------------------

df_sales['Total_Revenue'] = df_sales['Sales'] * df_sales['Quantity']

print("\n5.1 DataFrame with Total_Revenue column:")
print(df_sales.head())


# --------------------------------------------------
# 5.2 Sort the DataFrame
# First sort by Date in ascending order
# Then sort by Sales in descending order
# --------------------------------------------------

df5_2_sorted = df_sales.sort_values(
    by=['Date', 'Sales'],
    ascending=[True, False]
)

print("\n5.2 Sorted DataFrame:")
print(df5_2_sorted.head())


# --------------------------------------------------
# 5.3 Group by Region
# Calculate the total Sales and Quantity
# --------------------------------------------------

df5_3_grouped_region = df_sales.groupby('Region')[['Sales', 'Quantity']].sum()

print("\n5.3 Total Sales and Quantity by Region:")
print(df5_3_grouped_region)


# --------------------------------------------------
# 5.4 Group by Product
# Calculate the average Sales of each Product
# --------------------------------------------------

df5_4_grouped_product = df_sales.groupby('Product')['Sales'].mean()

print("\n5.4 Average Sales by Product:")
print(df5_4_grouped_product)
```
