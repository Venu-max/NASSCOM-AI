```python
#1.1 Generate a list of 100 simulated sensor readings.
import os
import datetime
import random

# Global Data
DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor', 'Pressure_Sensor', 'Light_Sensor']

LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

# List to store sensor data
raw_sensor_data = []

for i in range(1, 101):

    # Current timestamp
    timestamp_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Device ID
    device_id = f"DEV{i:03d}"

    # Random device type
    device_type = random.choice(DEVICE_TYPES)

    # Random location
    location = random.choice(LOCATIONS)

    # Generate random value based on device type
    low, high = READING_RANGES[device_type]
    value = round(random.uniform(low, high), 2)

    # Create tuple
    reading = (timestamp_str, device_id, device_type, location, value)

    # Add to list
    raw_sensor_data.append(reading)

# Print first 10 records
for record in raw_sensor_data[:10]:
    print(record)
```

```python
#1.2 From 'raw_sensor_data', create a list named 'temperature_readings' containing only the readings
temperature_readings = []

for reading in raw_sensor_data:
    timestamp, device_id, device_type, location, value = reading

    if device_type == "Temperature_Sensor":
        temp_dict = {
            "timestamp": timestamp,
            "device_id": device_id,
            "location": location,
            "temperature_celsius": value
        }
        temperature_readings.append(temp_dict)

# Print first 5 records
print(temperature_readings[:5])
```

```python
# 1.3 Create a set named 'unique_locations' containing all unique locations from 'raw_sensor_data'.
unique_locations = set()

for reading in raw_sensor_data:
    unique_locations.add(reading[3])

print(unique_locations)
```

```python
# 1.4 Create a dictionary named 'device_stats' where keys are 'device_type' and values are
device_stats = {}

# Initialize dictionary
for device in DEVICE_TYPES:
    device_stats[device] = (float('inf'), float('-inf'))

# Update min and max values
for reading in raw_sensor_data:
    device_type = reading[2]
    value = reading[4]

    min_val, max_val = device_stats[device_type]

    if value < min_val:
        min_val = value

    if value > max_val:
        max_val = value

    device_stats[device_type] = (min_val, max_val)

print(device_stats)
```

```python
##Exercise 2
def get_readings_by_location(readings_list, target_location):
    result = []

    for reading in readings_list:
        if reading[3] == target_location:
            result.append(reading)

    return result

# Example
print(get_readings_by_location(raw_sensor_data, "Lab1"))
```

```python
# 2.2 Write a function `calculate_average_temperature(temp_data_list)`
def calculate_average_temperature(temp_data_list):

    if len(temp_data_list) == 0:
        return 0.0

    total = 0

    for item in temp_data_list:
        total += item["temperature_celsius"]

    average = total / len(temp_data_list)

    return average

# Example
print(calculate_average_temperature(temperature_readings))
```

```python
#2.3 Write a function `generate_report_line(reading_dict)`
def generate_report_line(reading_dict):

    return f"[{reading_dict['timestamp']}] Device {reading_dict['device_id']} at {reading_dict['location']}: {reading_dict['temperature_celsius']}°C"

# Example
print(generate_report_line(temperature_readings[0]))
```

```python
# 3.2 Calculate the duration between the first and last timestamp in 'raw_sensor_data'.
if len(raw_sensor_data) > 1:

    first_timestamp_str = raw_sensor_data[0][0]
    last_timestamp_str = raw_sensor_data[-1][0]

    # Convert string to datetime object
    date_format = "%Y-%m-%d %H:%M:%S"

    first_dt = datetime.datetime.strptime(first_timestamp_str, date_format)
    last_dt = datetime.datetime.strptime(last_timestamp_str, date_format)

    duration = last_dt - first_dt

    days = duration.days
    seconds = duration.total_seconds()

    hours = int(seconds // 3600) % 24
    minutes = int((seconds % 3600) // 60)

    print(f"\nDuration of data collection: {days} days, {hours} hours, {minutes} minutes.")

else:
    print("\nNot enough data to calculate duration.")
```

```python
# Hands-on #1 vectors,dot products and Norms in Numpy
import numpy as np

u = np.array([2, 3])
v = np.array([4, 1])

print("u + v =", u + v)
print("3 * u =", 3 * u)
```

```python
dot1 = np.dot(u, v)

dot2 = u @ v

cos_theta = np.dot(u, v)/(np.linalg.norm(u)*np.linalg.norm(v))

theta = np.arccos(cos_theta)

print(f"Angle: {np.degrees(theta):.1f}°")
```

```python
w = np.array([3,-4])

print("L1:", np.linalg.norm(w,1))
print("L2:", np.linalg.norm(w,2))
print("Linf:", np.linalg.norm(w,np.inf))
```

```python
#Exercise
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

student1 = np.array([90,85,80])
student2 = np.array([88,84,79])
student3 = np.array([60,95,70])

print("Student1 L2:", np.linalg.norm(student1))
print("Student2 L2:", np.linalg.norm(student2))
print("Student3 L2:", np.linalg.norm(student3))

print(cosine_similarity([student1],[student2]))
print(cosine_similarity([student1],[student3]))
print(cosine_similarity([student2],[student3]))
```

```python
# Hands-on #2 Matrix Operations in NumPy
#Real -Life Scenario
#2 Pens + 1 Book + 1 Bag = ₹180
#1 Pen + 3 Books + 2 Bags = ₹370
#3 Pens + 2 Books + 1 Bag = ₹290
import numpy as np

A = np.array([
    [2,1,1],
    [1,3,2],
    [3,2,1]
])

b = np.array([180,370,290])

solution = np.linalg.solve(A,b)

print(solution)
```
